```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update({
        "font.weight": "bold",  # bold fonts
        "font.size":10,         #size of fonts
        "lines.linewidth": 2,   # thick lines
        "grid.color": "0.5",    # gray gridlines
        "grid.linewidth": 0.5,  # thin gridlines
        "savefig.dpi": 300,     # higher resolution output.
        "axes.labelsize" : 15, #size of label
        'legend.labelspacing':0.95, #spacing between labes
        'figure.figsize':(14,7) #figure size
    })
```

```python
metero_data = pd.read_csv("Data CSV/meterological_data.csv")
metero_data = metero_data.set_index('Date')
```

```python
metero_data.describe()
```

```python
#time series plots
fig,ax = plt.subplots()
metero_data.plot(ax=ax)

ax.legend()
```

```python
#scatter plot
temp = metero_data['avg_temp']
humid = metero_data['humidity']
rain = metero_data['avg_rain_mm']
```

```python
fig,ax = plt.subplots()
ax.scatter(temp,humid)
plt.savefig('scatter.jpeg',bbox_inches='tight')
```
